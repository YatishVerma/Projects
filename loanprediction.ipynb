```python
from sklearn.ensemble import IsolationForest
numeric_columns = df.select_dtypes(include='float64'or'int64')
skewness_values = numeric_columns.apply(lambda x: skew(x))
transformed_columns = numeric_columns.apply(lambda x: np.log1p(x) if skew(x) > 1 else (np.power(x, 1/3) if skew(x) < -1 else x))
isolation_forest_model = IsolationForest(contamination=0.05)
outlier_predictions = isolation_forest_model.fit_predict(transformed_columns)
outlier_mask = outlier_predictions == -1
data_no_outliers = df.loc[~outlier_mask, :]
```

```python
x=data_no_outliers.drop('Loan_Status',axis=1)
y=data_no_outliers['Loan_Status']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

numeric_features =numerical_variables 
categorical_features = categorical_variables
target_feature = 'Loan_Status'

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),  
        ('cat', OneHotEncoder(), categorical_features)  
    ])

label_encoder = LabelEncoder()

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
])


X_transformed = pipeline.fit_transform(x)
y_transformed = label_encoder.fit_transform(y)
print(X_transformed)
print(y_transformed)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=82)
X_resampled, y_resampled = smote.fit_resample(X_transformed, y_transformed)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X_resampled,y_resampled,random_state=44,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import roc_curve, auc, classification_report, confusion_matrix
from itertools import cycle
from sklearn.metrics import accuracy_score


models = [
    ('Logistic Regression', LogisticRegression()),
    ('Decision Tree', DecisionTreeClassifier()),
    ('Random Forest', RandomForestClassifier()),
    ('Support Vector Machine', SVC(probability=True))
]

# Plot ROC curves for each model
plt.figure(figsize=(12, 8))
for name, model in models:
    model.fit(x_train, y_train)
    y_prob = model.predict_proba(x_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Guessing')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Binary Classification Models')
plt.legend(loc='lower right')
plt.show()

# Classification reports and confusion matrices
for name, model in models:
    print(f"\nClassification Report and Confusion Matrix for {name}:\n")
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred)
    print(f"Confusion Matrix:\n{cm}\n")
    accuracy = accuracy_score(y_test, y_pred)

    print(f'Accuracy: {accuracy:.2f}')
```

```python
from sklearn.model_selection import cross_val_score

# Assuming X and y are your features and target variable
# Assuming models is the list of binary classification models

for name, model in models:
    # Use cross_val_score to perform 5-fold cross-validation (you can adjust the cv parameter)
    accuracy_scores = cross_val_score(model,X_resampled,y_resampled, cv=5, scoring='accuracy')

    # Print the accuracy scores for each fold and the average accuracy
    print(f"{name} - Cross-Validation Accuracy Scores: {accuracy_scores}")
    print(f"{name} - Average Accuracy: {accuracy_scores.mean():.2f}\n")
```

random forest have highesr accuracy 


```python
from sklearn.model_selection import GridSearchCV
param_grid={   
'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
rf_classifier = RandomForestClassifier()
grid_search = GridSearchCV(rf_classifier, param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)
print("Best Hyperparameters:", grid_search.best_params_)
best_model = grid_search.best_estimator_
accuracy = best_model.score(x_test, y_test)
print("Accuracy on Test Set:", accuracy)
```
