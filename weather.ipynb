```python
import numpy as np
import pandas as pd
from sklearn.model_selection import TimeSeriesSplit
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

y = df['Next_Tmax']

order = (1, 1, 1)  

tscv = TimeSeriesSplit(n_splits=5)  

for train_index, test_index in tscv.split(y):
    train_data, test_data = y.iloc[train_index], y.iloc[test_index]

    model = ARIMA(train_data, order=order)
    fitted_model = model.fit()

    forecast_values = fitted_model.forecast(steps=len(test_data))

    mse = mean_squared_error(test_data, forecast_values)
    print(f"Mean Squared Error: {mse}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error



X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree Regressor': DecisionTreeRegressor(),
    'Random Forest Regressor': RandomForestRegressor(),
    'Support Vector Regressor': SVR()
}

for model_name, model in models.items():
    model.fit(X_train, y_train)

    predictions = model.predict(X_test)

    mse = mean_squared_error(y_test, predictions)
    print(f"{model_name} Mean Squared Error: {mse}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.decomposition import PCA


rf_model = RandomForestRegressor()
rf_cv_scores = cross_val_score(rf_model, X_pca, y, cv=5, scoring='neg_mean_squared_error')

svr_model = SVR()
svr_cv_scores = cross_val_score(svr_model, X_pca, y, cv=5, scoring='neg_mean_squared_error')

print(f"Random Forest CV Mean Squared Error: {-np.mean(rf_cv_scores)}")
print(f"Support Vector Regressor CV Mean Squared Error: {-np.mean(svr_cv_scores)}")
```

```python
import numpy as np
from sklearn.svm import SVR
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error


X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

svr_model = SVR()

param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'epsilon': [0.1, 0.2, 0.5],
    'gamma': ['scale', 'auto']
}

grid_search = GridSearchCV(svr_model, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

best_svr_model = grid_search.best_estimator_

predictions = best_svr_model.predict(X_test)

mse = mean_squared_error(y_test, predictions)
print(f"SVR Mean Squared Error on Test Set: {mse}")
```
